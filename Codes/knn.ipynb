```python
df1.columns
```

```python
def grade_label(x):
    if x==1:
        return 2
    elif x==0.5:
        return 1
    else:
        return 0

df1['Grade'] = df1['Grade'].apply(lambda x : grade_label(x))

print("Target variable Grade after encoding ")
df1
```

```python
x=df1.drop(['Grade'],axis=1)
x.head()
```

```python
y=df1['Grade']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(xtrain,ytrain)
```

```python
ypred=knn.predict(xtest)

ypred
```

```python
ytest
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
```

```python
error_rate=[]

for i in range(1,50):
    knn=KNeighborsClassifier(n_neighbors=i)
    
    knn.fit(xtrain,ytrain)
    pred_i=knn.predict(xtest)
    
    error_rate.append(np.mean(pred_i != ytest))
    
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,50),error_rate,color='blue',linestyle='--',markersize=10,markerfacecolor='red',marker='o')
plt.title('K versus Error rate')
plt.xlabel('K')
plt.ylabel('Error rate')

```

```python
# error is lowest at 1,2,4,5 ,so better to take k value is 5 to get more accuracy

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(xtrain,ytrain)
```

```python
predictions=knn.predict(xtest)

print(confusion_matrix(ytest,predictions))
print(classification_report(ytest,predictions))
```
